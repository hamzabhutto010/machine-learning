```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

##  Scatterd plot numerical - numerical 

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

```python
train = pd.read_csv('train.csv')
train
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

### bar plot ( numerical - categorical)

```python

sns.barplot(x = train['Pclass'], y = train['Age'], hue=train['Sex'] )
```

### heatmap (catagorical and catagorical)

```python
sns.heatmap(pd.crosstab(train['Pclass'],train['Survived']))
```

```python
pd.crosstab(train['Pclass'], train['Survived'])
```

```python
# heat  map
sns.clustermap(pd.crosstab(train['Pclass'], train['Survived']))
```

```python
# pair plot 
sns.pairplot(train, hue='Sex')
```
