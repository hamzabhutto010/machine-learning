```python
import pickle 
import numpy as np
```

```python
ohe_sex = pickle.load(open('model/ohe_sex.pk1', 'rb'))
ohe_emb = pickle.load(open('model/ohe_embarked.pk1', 'rb'))
clf = pickle.load(open('model/clf.pk1', 'rb'))
```

```python
# Predicting using the saved model
# test input
test_input = np.array([2, 'male', 31.0, 0, 0, 10.5, 'S'],dtype=object).reshape(1,7)
```

```python
test_input_sex = ohe_sex.transform(test_input[:,1].reshape(-1,1))
test_input_emb = ohe_emb.transform(test_input[:,6].reshape(-1,1))
test_input_final = np.hstack([test_input[:,[0,2,3,4,5]], test_input_sex, test_input_emb])
```

```python
test_input_final.shape
```

```python
if clf.predict(test_input_final) == 1:
    print("Survived")
elif clf.predict(test_input_final) == 0:   
    print('die')
```
