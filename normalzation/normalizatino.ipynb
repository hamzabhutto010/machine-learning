```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine = pd.read_csv('wine_data.csv', usecols=[0,1,2], header=None)
wine.head(3)
```

```python
wine.columns=['Class label', 'Alcohol', 'Malic acid']
```

```python
wine.sample(5)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(wine.drop('Class label', axis=1), wine['Class label'], test_size=0.2, random_state=42)
```

```python
x_train.shape
x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scale = MinMaxScaler()

# it will learn the parameter and data distribution
scale.fit(x_train)

# it will laern transform the value to apply the formula of min max scaling
scale_x_train = scale.transform(x_train)
scale_x_test = scale.transform(x_test)

```

```python
scale_x_train = pd.DataFrame(scale_x_train, columns=x_train.columns)
scale_x_test = pd.DataFrame(scale_x_test, columns=x_test.columns)
```
