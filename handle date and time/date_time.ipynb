```python
import pandas as pd 
import numpy as np
```

```python
date = pd.read_csv('orders.csv')
time = pd.read_csv('messages.csv')
```

```python
date.head()
```

## working with Date

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
date['year'] = date['date'].dt.year
date.head(2)
```

```python
date['month'] = date['date'].dt.month_name()
```

```python
date['day'] = date['date'].dt.day
```

```python
date['week'] = date['date'].dt.dayofweek
```

```python
date['quarter'] = date['date'].dt.quarter
```

```python
date['day_name'] = date['date'].dt.day_name()
```

```python
date
```

```python
date['holiday'] = np.where(date['day_name'].isin(['Sunday', 'Saturday']), 1, 0)
```
