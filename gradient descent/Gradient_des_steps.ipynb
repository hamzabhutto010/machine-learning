```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
plt.scatter(x,y)
plt.show()
```

## let applty OLS (linear regression)

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.plot(x,lr.predict(x),color='red')
plt.scatter(x,y)
```

## now apply the Gradient descent ::
assuming the slope is constant m = 78
and let start the intercept value b = 0

```python
y_pred = ((78.35 * x) + 0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='blue',label='ols')
plt.plot(x,y_pred,color='red',label='b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 100
loss_slope = -2 * np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
# let learning rate = 0.1
learning_rate = 0.1
step_size = loss_slope*learning_rate
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='blue',label='ols')
plt.plot(x,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(x,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# iteration = 2
loss_slope = -2 * np.sum(y - m*x.ravel() - b)
step_size = loss_slope*learning_rate
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='blue',label='ols')
plt.plot(x,y_pred,color='#00a65a',label='b = {}'.format(b))
plt.plot(x,y_pred1,color='#A3E4D7',label='b = 0')
plt.plot(x,y_pred2,color='black',label='b = {}'.format(b))
plt.legend()
plt.show()
```

```python
# initail random value of b = -100
b = -100
m = 78.35
lr1 = 0.01

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*x.ravel() - b)
  step_size = loss_slope*lr1
  b = b - step_size
  y_pred = ((78.35 * x) + b).reshape(4)
  plt.plot(x,y_pred,)
plt.scatter(x,y)
```
