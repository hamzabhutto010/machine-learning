```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tip = pd.read_csv('tips.csv')
tip
```

```python
# find the female and male 
sns.countplot(x='sex', data= tip)
```

```python
# find the smoker
sns.countplot(x= 'smoker', data = tip)
```

```python
# find the time 
tip['time'].value_counts().plot(kind='pie', autopct='%2f')
```

```python
# tip
plt.hist(x='tip', data = tip, bins=10)
```

```python
sns.distplot(tip['total_bill'])
```

```python
sns.boxplot(tip['total_bill'])
```
