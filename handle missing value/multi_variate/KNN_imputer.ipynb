```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer
```

```python
df = pd.read_csv('train.csv', usecols=['Fare','Age','Survived'])
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
x = df[['Fare','Age']]
y = df['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
imputer = KNNImputer(n_neighbors=3, weights='distance')
x_train_imputed = imputer.fit_transform(x_train)
x_test_imputed = imputer.transform(x_test)
```

```python
x_train_imputed = pd.DataFrame(x_train_imputed, columns=x_train.columns)
```

```python
x_test_imputed = pd.DataFrame(x_test_imputed, columns=x_test.columns)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()

clf.fit(x_train_imputed, y_train)
prd = clf.predict(x_test_imputed)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, prd)
```

```python
# simple mean imputer

from sklearn.impute import SimpleImputer


imputer = SimpleImputer(strategy='mean')
x_train_imputed2 = imputer.fit_transform(x_train)
x_test_imputed2 = imputer.transform(x_test)

clf.fit(x_train_imputed2, y_train)
pred2 = clf.predict(x_test_imputed2)
accuracy_score(y_test, pred2)
```
